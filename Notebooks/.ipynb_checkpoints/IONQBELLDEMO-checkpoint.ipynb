```python
ionq_token="LAYq6EwdvXqh4sIma2iKZNQ2l7KzbxX6"
from qiskit_ionq import IonQProvider
provider = IonQProvider(ionq_token)
backend_qpu = provider.get_backend("ionq_qpu")
```

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2,2)   
qc.h(0)  
qc.cx(0,1) 
qc.measure([0,1],[0,1]) 
qc.draw(output='mpl')
```

```python
#run on IonQ hardware:  
qpu_job = backend_qpu.run(qc, shots=50000)  #We run 50000 shots of the Bell circuit

# since the job is queued and may take several minutes to get processed, 
# we check its status every minute until the job is done
from qiskit.providers.jobstatus import JobStatus
import time
while qpu_job.status() is not JobStatus.DONE:
   print("Job status is", qpu_job.status())
   time.sleep(60)
# break out of the loop when the job finished:
print("Job status is",qpu_job.status())

counts = qpu_job.get_counts()
print("IonQ hardware: measured counts",counts)
print("IonQ hardware: measured probabilities",qpu_job.get_probabilities())
```

```python
# plot the results
all_states = ['00','01','10','11']
if len(counts) < 4:
    for j in all_states:
        if j not in counts.keys():
            counts[j]=0
reordered = {k:counts[k] for k in all_states}

from qiskit.visualization import plot_histogram
plot_histogram(reordered)
```

```python
#Lets try generating bell state |00> + |11> , with both qubits having bit-flip protection now.

from qiskit import QuantumCircuit

qc = QuantumCircuit(6,2)   
#First encode each logical qubit (0,1) into 3 qubits (0 2 3, 1 4 5)
qc.cx(0,2)
qc.cx(0,3)

qc.cx(1,4)
qc.cx(1,5)

#Next introducing potential error
qc.h(0)
qc.cx(0,1)

#Next decoding/error correction
qc.cx(0,2)
qc.cx(0,3)
qc.cx(1,4)
qc.cx(1,5)

qc.ccx(3,2,0)
qc.ccx(5,4,1)

qc.measure([0,1],[0,1]) 
qc.draw(output='mpl')
```

```python
#run on IonQ hardware:  
qpu_job = backend_qpu.run(qc, shots=10000)  #We run 10000 shots of the error corrected Hadamard circuit

# since the job is queued and may take several minutes to get processed, 
# we check its status every minute until the job is done
from qiskit.providers.jobstatus import JobStatus
import time
while qpu_job.status() is not JobStatus.DONE:
   print("Job status is", qpu_job.status())
   time.sleep(60)
# break out of the loop when the job finished:
print("Job status is",qpu_job.status())

counts = qpu_job.get_counts()
print("IonQ hardware: measured counts",counts)
print("IonQ hardware: measured probabilities",qpu_job.get_probabilities())
```

```python
# plot the results
all_states = ['00','01','10','11']
if len(counts) < 4:
    for j in all_states:
        if j not in counts.keys():
            counts[j]=0
reordered = {k:counts[k] for k in all_states}

from qiskit.visualization import plot_histogram
plot_histogram(reordered)
```
